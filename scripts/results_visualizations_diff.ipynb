```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
plt.style.use(['seaborn-v0_8','voresStyle.mplstyle'])
```

```python
# Read the CSV file into a DataFrame
plt.figure(figsize=(8, 3))

for i, file in enumerate(os.listdir("../results/diff_csv")):
    if file.endswith(".csv"):
        df = pd.read_csv('../results/diff_csv/' + file, header=0, usecols=range(19))
        plt.semilogy(df['step'], df['mse'], label=f'Model {i} - MSE')
        #plt.plot(df['step'], df['mse'], label=f'Model {i} - MSE')
plt.grid(True, which="both", ls="-")
plt.xlabel('Step')
plt.ylabel('Log-transformed MSE Loss')
plt.yscale('log')
plt.title('Diffusion models: Train MSE Loss over Steps')
plt.legend()
plt.savefig('../results/diff_csv/mse_loss.pdf', bbox_inches='tight')
plt.show()
```

```python
# Read the CSV file into a DataFrame
plt.figure(figsize=(8, 3))

for i, file in enumerate(os.listdir("../results/diff_csv")):
    if file.endswith(".csv"):
        df = pd.read_csv('../results/diff_csv/' + file, header=0, usecols=range(19))
        plt.semilogy(df['step'], df['loss'], label=f'Model {i} - Loss')
plt.grid(True, which="both", ls="-")
plt.xlabel('Step')
plt.ylabel('Log-transformed Loss')
plt.yscale('log')
plt.title('Diffusion models: Train Loss over Steps')
plt.legend()
plt.savefig('../results/diff_csv/loss.pdf', bbox_inches='tight')
plt.show()
```
