```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('../results/Diff-2024-04-09-16-24-50-133538/progress.csv', header=0, usecols=range(19))
```

```python
df.head()
```

```python
# Visualize grad_norm and param_norm columns

sns.pairplot(df[['grad_norm', 'param_norm']])
plt.show()
```

```python

# Visualize samples column
plt.figure(figsize=(10, 6))
plt.plot(df['step'],df['samples'], label='samples')
plt.xlabel('Step')
plt.ylabel('Samples')
plt.title('Samples over Steps')
plt.legend()
plt.show()
```

```python

# Visualize train_acc@1 and train_acc@2 columns
plt.figure(figsize=(10, 6))
plt.plot(df['step'], df['loss'], label='loss')
plt.plot(df['step'], df['mse'], label='mse')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Train Loss over Steps')
plt.legend()
plt.show()
```

```python

# Visualize train_loss column
plt.figure(figsize=(10, 6))
plt.plot(df['step'], df['vb'], label='vb')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Train Loss over Steps')
plt.legend()
plt.show()
```
