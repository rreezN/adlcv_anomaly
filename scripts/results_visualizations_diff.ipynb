```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
plt.style.use(['seaborn-v0_8','voresStyle.mplstyle'])
```

```python
# Read the CSV file into a DataFrame
plt.figure(figsize=(10, 6))

for i, file in enumerate(os.listdir("../results/diff_csv")):
    if file.endswith(".csv"):
        df = pd.read_csv('../results/diff_csv/' + file, header=0, usecols=range(19))
        plt.plot(df['step'], np.log(df['mse']), label=f'Model {i+1} - mse')

plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Train Loss over Steps')
plt.legend()
plt.savefig('../results/diff_csv/mse_loss.pdf')
plt.show()
#df = pd.read_csv('../results/Diff-2024-04-09-16-24-50-133538/progress.csv', header=0, usecols=range(19))
```
