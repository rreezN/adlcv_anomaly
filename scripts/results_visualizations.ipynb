```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('../results/Class-2024-04-09-16-59-58-670362/progress.csv', header=0, usecols=range(19))
```

```python
df.head()
```

```python
# Visualize grad_norm and param_norm columns

sns.pairplot(df[['grad_norm', 'param_norm']])
plt.show()
```

```python

# Visualize samples column
plt.figure(figsize=(10, 6))
plt.plot(df['step'],df['train_loss'], label='Train Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Train loss over Steps')
plt.legend()
plt.show()
```

```python

# Visualize train_acc@1 and train_acc@2 columns
plt.figure(figsize=(10, 6))
plt.plot(df['step'], df['train_acc@1'], label='train_acc@1')
plt.plot(df['step'], df['train_acc@2'], label='train_acc@2')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Train Accuracy over Steps')
plt.legend()
plt.show()
```

```python

# Visualize train_loss column
plt.figure(figsize=(10, 6))
plt.plot(df['step'], df['train_loss'], label='train_loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Train Loss over Steps')
plt.legend()
plt.show()
```
